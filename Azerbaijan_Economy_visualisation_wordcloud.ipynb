# Azerbaijan Economy words cloud

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
```

```python
# install wordcloud
!conda install -c conda-forge wordcloud==1.4.1 --yes

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

##### You can get the information for word cloud from Wikipedia page of **Azerbaijan Economy**. And then just I save it as Cntrl+A & Cntrl+C to 'txt' extensioned file

```python
# open the file and read it into a variable azerbaijan_economy
azerbaijan_economy = open('azerbaijan_economy.txt', 'r').read()
    
print ('File downloaded and saved!')
```

```python
# for deleting redundant words
stopwords = set(STOPWORDS)
```

```python
# creating cloud of important 2000 words
# instantiate a word cloud object
azerbaijan_wc = WordCloud(
    background_color='white',
    max_words=5000,
    stopwords=stopwords
)

# generate the word cloud
azerbaijan_wc.generate(azerbaijan_economy)
```

```python
# now time to visualise
plt.imshow(azerbaijan_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# for better view
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(azerbaijan_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### It is better to delete "months" and other less informative words such as "web", "org", "pdf" from our beautiful cloud as they aren't important words for us

```python
stopwords.add('January')
stopwords.add('February')
stopwords.add('March')
stopwords.add('April')
stopwords.add('May')
stopwords.add('June')
stopwords.add('July')
stopwords.add('August')
stopwords.add('September')
stopwords.add('October')
stopwords.add('November')
stopwords.add('December')
stopwords.add('org')
stopwords.add('web')
stopwords.add('pdf')
stopwords.add('Bln')
stopwords.add('est')
stopwords.add('Retrieved')
stopwords.add('Archived')

# re-generate the word cloud
azerbaijan_wc.generate(azerbaijan_economy)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(azerbaijan_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Now time to save Azerbaijan map

First of all I download geographical transparent map of Azerbaijan from https://vemaps.com/azerbaijan/az-01

Then I filled in with black and out with white in Adobe Photoshop

Then import to my lab under 'az-map_new.png'

```python
# save mask to alice_mask
azerbaijan_map = np.array(Image.open('az-map_new.png'))
    
print('Image downloaded and saved!')
```

#### Let's look what we downloaded

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(azerbaijan_map, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Let's fill it with our words

```python
# instantiate a word cloud object
azerbaijan_wc = WordCloud(background_color='white', max_words=2000, mask=azerbaijan_map, stopwords=stopwords)

# generate the word cloud
azerbaijan_wc.generate(azerbaijan_economy)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(azerbaijan_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

And We did it!
Now we have word cloud map with Azerbaijan map
